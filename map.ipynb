```python
def find_path(start, end):
    # Find the distance between two points.
    def distance(point1, point2):
        return (point1[0] - point2[0])**2 + (point1[1] - point2[1])**2

    # Find the point in a list of points that is closest to a target point.
    def point_nearest_to_target(points, target):
        return min(points, key=lambda point: distance(point, target))

    # Find the closest intersection to the start point.
    new_start = point_nearest_to_target(intersections, start)
    # Find the closest intersection to the end point.
    new_end = point_nearest_to_target(intersections, end)

    # Trace a route from the new start point to the new end point.
    route = [new_start]

    # Try to find a route from the current point to the new end point.
    def find_next_point(cur_point, end):
        # Find the points adjacent to the current point that have not yet been visited.
        adjacents_not_visited = [
            adjacent for adjacent in adjacents[hash(cur_point)] if adjacent not in route]
        # Sort the adjacent points by their distance to the new end point.
        for point in sorted(adjacents_not_visited, key=lambda adjacent: distance(adjacent, end)):
            route.append(point)
            if point == end:
                raise Exception("Route found")
            else:
                find_next_point(point, end)
            route.remove(point)

    try:
        if new_start == new_end:
            raise Exception("Route found")
        find_next_point(new_start, new_end)
    except Exception as exception:
        if "Route found" in str(exception):
            return [start] + route + [end]
        else:
            raise exception
```

Chuong trinh Final

```python
#Tạo layer group để vẽ
layer_group = LayerGroup()
map.add_layer(layer_group)

# Khởi tạo start_marker, end_marker, polyline giá trị null
start_marker = None
end_marker = None
polyline = None


# Tạo marker bằng việc click vào bản đồ, đồng thời vẽ polyline
def handle_interaction_marker(**kwargs):
    global start_marker, end_marker
    if kwargs.get('type') == 'click':
        if start_marker is None:
            start_marker = Marker(location=kwargs.get('coordinates'), set_color = 'red', draggable=True, title="Start")
            # start_icon = Icon(icon='fa-circle', color='red', icon_color='white')
            # start_marker.icon = start_icon
            # start_marker.icon_color = 'red'
            start_marker.on_move(handle_move_start_marker)
            start_marker.on_click(handle_onclick_marker)
            layer_group.add_layer(start_marker)
            map.add_layer(start_marker)
        elif end_marker is None:
            end_marker = Marker(location=kwargs.get('coordinates'), draggable=True, title="End")
            end_marker.on_move(handle_move_end_marker)
            end_marker.on_click(handle_onclick_marker)
            layer_group.add_layer(end_marker)
            map.add_layer(end_marker)
            # Vẽ polyline khi đã biết được tọa độ 2 marker
            draw_path(start_marker.location, end_marker.location)


# Xóa marker đã tạo ( xóa start_marker thì xóa cả end_marker, còn nếu xóa end_marker thì vẫn giữ start_marker )
def handle_onclick_marker(**kwargs):
    global start_marker, end_marker
    if kwargs.get('type') == 'click':
        for layer in layer_group.layers:
            if layer.location == kwargs['coordinates']:
                map.remove_layer(layer)
                layer_group.remove_layer(layer)
                if layer == start_marker:
                    start_marker = None
                    clear_path()
                    if end_marker is not None:
                        map.remove_layer(end_marker)
                        layer_group.remove_layer(end_marker)
                        end_marker = None
                elif layer == end_marker:
                    end_marker = None
                    clear_path()


# Di chuyển start_marker
def handle_move_start_marker(**kwargs):
    draw_path(kwargs["location"], end_marker.location)

# Di chuyển end_marker
def handle_move_end_marker(**kwargs):
    draw_path(start_marker.location, kwargs["location"])


# Tạo đường đi polyline từ start_marker đến end_marker
def draw_path(start_location, end_location):
    global polyline
    if start_marker is not None and end_marker is not None:
        if polyline is not None:
            map.remove_layer(polyline)
        polyline = AntPath(locations=find_path(start_location, end_location), fill=False) 
            #dùng find_path chạy đc 20 lần move marker ( move 2 cả start lẫn end)
            #chuyển sang đường polyline thẳng thì thay find_path(start_location, end_location) bằng [start_location, end_location]
        map.add_layer(polyline)
        
        
        # try:
        #     polyline = AntPath(locations=find_path(start_location, end_location), fill=False) 
        #     #dùng find_path chạy đc 10 lần move marker ( move 2 cả start lẫn end)
        #     #chuyển sang đường polyline thẳng thì thay find_path(start_location, end_location) bằng [start_location, end_location]
        #     map.add_layer(polyline)
        # except TraitError:
        #     print("No route found!")
        


# Xóa đường polyline đang tồn tại
def clear_path():
    global polyline
    if polyline is not None:
        map.remove_layer(polyline)
        polyline = None


map.on_interaction(handle_interaction_marker)
display(map)
```
